```python
members_churn_log = pd.merge(train, members, on="msno", how="left")
members_churn_log.columns
non_msno = members_churn_log.drop(["msno"], axis=1)
```

```python
non_msno.gender.unique()
```

```python
category_mapping = {'female': 0, 'male': 1}

# Use the map method to create a new column with encoded values
non_msno['gender'] = non_msno['gender'].map(category_mapping)
non_msno.gender.unique()
```

```python
corr_matrix = non_msno.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
